```python
import numpy as np
from scipy.stats import norm
```

```python
def Black_Scholes_Pricing(S0,K,T,r,delta,sigma):
    d1 = (np.log(S0/K) + (r - delta + sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma* np.sqrt(T)

    call = S0 * np.exp(-delta*T) * norm.cdf(d1) - K * np.exp(-r*T)*norm.cdf(d2)
    return call
```

```python
BS_price = Black_Scholes_Pricing(6849.09,6750,2.052,0.0339,0.0114,0.2467)
print(f"The BS pricing for this option is: {BS_price:.2f}")
```
